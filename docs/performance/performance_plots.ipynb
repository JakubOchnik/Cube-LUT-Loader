```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
TRI_DATA_PATH = 'data/trilinear.csv'
NV_DATA_PATH = 'data/nearest_val.csv'

TRI_GRAPH_PATH = 'img/tri_interp.png'
NV_GRAPH_PATH = 'img/nv_interp.png'
```

```python
trilinear = pd.read_csv(TRI_DATA_PATH, index_col=0)
nv = pd.read_csv(NV_DATA_PATH, index_col=0)
```

## Performance comparison of trilinear interpolation ran on a CPU and a GPU

```python
trilinear.plot(use_index=True, y=["1 thread", "8 threads", "GPU"],
         figsize=(10, 10), logy=True)
plt.title("Performance comparison (trilinear interpolation)")
plt.xlabel('Image size')
plt.ylabel('Time (s)')
plt.grid(which='both')
plt.savefig(TRI_GRAPH_PATH, dpi=80, facecolor='white', bbox_inches='tight')
```

## Performance comparison of nearest-value interpolation ran on a CPU and a GPU

```python
nv.plot(use_index=True, y=["1 thread", "8 threads", "GPU"],
         figsize=(10, 10), logy=True)
plt.title("Performance comparison (nearest-value interpolation)")
plt.xlabel('Image size')
plt.ylabel('Time (s)')
plt.grid(which='both')
plt.savefig(NV_GRAPH_PATH, dpi=80, facecolor='white', bbox_inches='tight')
```
